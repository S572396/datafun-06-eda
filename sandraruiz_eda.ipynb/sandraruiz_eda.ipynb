## 5. Data Transformation and Feature Engineering, Exploring More Data

### 1. Renaming a Column

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Load the diamonds dataset into a pandas DataFrame
df = sns.load_dataset('diamonds')

# Rename columns 'x', 'y', and 'z' to 'length', 'width', and 'depth'
df = df.rename(columns={'x': 'length', 'y': 'width', 'z': 'depth'})

# Print column names
print("Column names after renaming:")
print(df.columns)

# Display the first 4 rows of the DataFrame
print("\nFirst 4 rows of the DataFrame:")
print(df.head(4))

```

* <b> Insights: Columns originally named, x,y, and z have been sucessfully renamed to length, width, and depth for the diamond data set. This makes the information easier to undstand. 

### 2.Adding in a new column named best_value:

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Load the diamonds dataset into a pandas DataFrame
df = sns.load_dataset('diamonds')

# Create a new column 'best_value' based on specified conditions
conditions = (
    (df['price'] < 5000) &
    (df['color'].isin(['D', 'E', 'F'])) &
    (df['cut'].isin(['Ideal', 'Premium'])) &
    (df['clarity'].isin(['VS1', 'VS2', 'IF']))
)
df['best_value'] = conditions.astype(int)

# Filter rows where 'best_value' is 1
best_value_rows = df[df['best_value'] == 1]

# Display the DataFrame with 'best_value' equal to 1
print("Rows with 'best_value' equal to 1:")
print(best_value_rows)

# Count the total number of rows that meet the 'best_value' specification
total_best_value_count = df['best_value'].sum()


```

* <b> Insights: A new column named best_value was sucessfuly created to add to the data set. The results show a total of 5,967 diamonds meet the specified critia for price less than 5,000 ; color in the D, E, or F category; cut in the Ideal or Premium category, and clarity set to include only VS1, VS2 or IF values.

## 6. Let's look at how many diamonds in this data set are considered prize diamonds, meaning they meet the criteria of the established 4 C's of cut, color, clarity and carat.

```python
# Filter the DataFrame based on the specified conditions
filtered_df = df[(df['cut'] == 'Ideal') & (df['color'] == 'D') & (df['clarity'] == 'IF') & (df['carat'] > 0.8)]

# Get the count of rows that meet all the criteria
count_best_values = len(filtered_df)

# Print the result
print("Number of diamonds meeting the best values:", count_best_values)
```

*  <b> Answer: The findings show that out of this large dataset of 53,940 diamonds only a mere 7 meet the criteria for best marks in each catory column of cut, color, clarity, and carat!!!  Where carat is >0.8, cut is the best at 'Ideal', color is the best at 'D', and clarity is the best at "IF" for Flawless.

## Let's Visualize the Data with a Scatter Plot: Price and Carat

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Sort the DataFrame by 'price' column in descending order
sorted_df = df.sort_values(by='price', ascending=False)

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=sorted_df)

# Set plot labels and title
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Scatter Plot of Diamonds - Price vs. Carat')

plt.show()
```

* <b> Insights: From the Scatter Plot we can see that carats at 1 or less do not reach high prices of 17,000 or greater. Diamond carats in the range of greater than 1 to 5 reach much higher prices, some as high as 18,823.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset into a pandas DataFrame
df = sns.load_dataset('diamonds')

# Create histograms for color, and clarity
plt.figure(figsize=(8, 16))


plt.subplot(4, 1, 2)
sns.histplot(x='price', data=df, hue='color', multiple='stack', palette='viridis', bins=30)
plt.title('Diamond Price Histogram by Color')

plt.subplot(4, 1, 3)
sns.histplot(x='price', data=df, hue='clarity', multiple='stack', palette='viridis', bins=30)
plt.title('Diamond Price Histogram by Clarity')


plt.tight_layout()
plt.show()
```
